```python
import tensorflow as tf
```

```python
import keras
from keras._tf_keras.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen1 = ImageDataGenerator(rescale = 1./255,shear_range = 0.2,zoom_range=0.2,horizontal_flip=True)
```

```python
!unzip "/content/pinkysiddu.zip" -d "/usr/local"
```

```python
training_set1 = train_datagen1.flow_from_directory('/usr/local/pinky/Animal Classification/dataset',target_size = (64,64),batch_size = 32,class_mode = 'categorical')
```

```python
test_datagen1 = ImageDataGenerator(rescale=1./255)
```

```python
cnn1 = tf.keras.models.Sequential()
```

```python
cnn1.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn1.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
```

```python
cnn1.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn1.add(tf.keras.layers.Flatten())
```

```python
cnn1.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
cnn1.add(tf.keras.layers.Dense(units=15,activation='softmax'))
```

```python
cnn1.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
cnn1.fit(x=training_set1,validation_data=test_set1,epochs=25)
```
